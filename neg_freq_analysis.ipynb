```python
# Necessary imports
import pandas as pd
import matplotlib.pyplot as plt
import ast
from matplotlib.ticker import MaxNLocator
```

Throughout this notebook the imaginary frequencies are referred to as negative frequencies since that's how they're distinguished as imaginary in the output files of the computational programs.

```python
# Make dataframe of xTB negative frequencies
df_xtb = pd.read_csv("C:/Users/sophi/OneDrive - University of Leeds/Year 4/Sophie Blanch/code/test_analysis/negative_freqs/negative_frequencies_100_random/xtb_negative_frequencies.csv")
df_xtb
```

```python
# Make xTB negative frequency histogram
df_xtb['File'] = df_xtb['File'].str.replace('.out', '', regex=False)

df_xtb['Negative Frequencies'] = pd.to_numeric(df_xtb['Negative Frequencies'], errors='coerce')

all_numbers = df_xtb['Negative Frequencies'].abs()



plt.figure(figsize=(8, 5), dpi=1000)

plt.hist(all_numbers, color='teal', edgecolor='black', alpha=0.7)

plt.xlabel('Vibrational Frequencies / cm')
plt.ylabel('Frequency')
plt.title('Imaginary Frequencies for 100 Random\nxTB Calculated Structures')
plt.gca().yaxis.set_major_locator(MaxNLocator(integer=True))
plt.rcParams['font.family'] = 'Times New Roman'
plt.savefig("C:/Users/sophi/OneDrive - University of Leeds/Year 4/Sophie Blanch/Report/images/xtb_imaginary_frequencies.png", dpi=1000, bbox_inches='tight')

plt.show()
```

```python
# Make PM6 negative frequency dataframe
df_pm6 = pd.read_csv("C:/Users/sophi/OneDrive - University of Leeds/Year 4/Sophie Blanch/code/test_analysis/negative_freqs/negative_frequencies_100_random/pm6_negative_frequencies.csv")
df_pm6
```

```python
# Modify PM6 negative frequency dataframe
df_pm6['File'] = df_pm6['File'].str.replace('.aux', '', regex=False)

df_pm6['Negative Frequencies'] = df_pm6['Negative Frequencies'].apply(ast.literal_eval)

df_pm6['Negative Frequencies'] = df_pm6['Negative Frequencies'].apply(lambda x: [float(i) for i in x])

df_pm6
```

```python
# Make PM6 negative frequency histogram
all_numbers = [abs(num) for sublist in df_pm6['Negative Frequencies'] for num in sublist]


plt.figure(figsize=(8, 5), dpi=1000)
plt.hist(all_numbers, bins=25, color='teal', edgecolor='black', alpha=0.7)
plt.xlabel('Vibrational Frequencies / cm')
plt.ylabel('Frequency')
plt.title('Imaginary Frequencies for 100 Random\nPM6 Calculated Structures')
plt.rcParams['font.family'] = 'Times New Roman'
plt.savefig("C:/Users/sophi/OneDrive - University of Leeds/Year 4/Sophie Blanch/Report/images/pm6_negative_frequencies.png", dpi=1000, bbox_inches='tight')

plt.show()
```

```python
# Add number of negative frequency column to DataFrame
df_xtb['Number of Negative Frequencies'] = 1
df_xtb
```

```python
# Add number of negative frequency column to DataFrame
df_pm6['Number of Negative Frequencies'] = df_pm6['Negative Frequencies'].apply(len)
df_pm6
```

```python
# Join DataFrames
merged_df = pd.merge(df_xtb, df_pm6, on='File', how='outer')
merged_df
```

```python
# Plot number of imaginary frequencies for PM6
no_of_imaginary_freqs = merged_df.drop(['Negative Frequencies_x', 'Negative Frequencies_y'], axis=1)

plt.figure(figsize=(8, 5), dpi=1000)
plt.hist(no_of_imaginary_freqs['Number of Negative Frequencies_y'], bins=25, color='blue', edgecolor='black', alpha=0.5)
plt.xlabel('Number of Imaginary Frequencies per Complex')
plt.ylabel('Frequency')
plt.title('Number of Imaginary Frequencies for 100 Random\nPM6 Calculated Structures')
plt.rcParams['font.family'] = 'Times New Roman'
plt.savefig("C:/Users/sophi/OneDrive - University of Leeds/Year 4/Sophie Blanch/Report/images/number_of_pm6_negative_frequencies.png", dpi=1000, bbox_inches='tight')

plt.show()
```
